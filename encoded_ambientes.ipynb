```python

df_filmes['Released_Year'] = pd.to_numeric(df_filmes['Released_Year'], errors='coerce')
#df_filmes['Released_Year'] = df_filmes['Released_Year'].apply(lambda x: int(x)) #--> Era pra essa função conseguir transformar todos os itens da coluna 'Released_Year' em inteiro
soma = 0
for i in df_filmes['Released_Year']:
  if i >= 2005 and i <= 2010:
    soma = soma + 1
print(soma)

df_filmes.info()
```
